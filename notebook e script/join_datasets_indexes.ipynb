```python
import pandas as pd
```

```python
df = pd.read_csv('out_citations_and_conferences_location_ready.csv',usecols=['ConferenceLocation'])

# vedo in quali Stati sono state svolte le conferenze 
countries = df['ConferenceLocation'].str.split(',').str[-1].str.lstrip().unique()
countries.sort()
print(countries)
print(len(countries))
```

```python
# Indici del WEF. devo tenermi soltanto quelli i cui Stati sono nel dataset delle conferenze
df_countries_WEF = pd.read_csv('wef_results.csv')
df_countries_WEF.sort_values(by=['Stato'], inplace=True)
list(df_countries_WEF['Stato'])
```

```python
# dataset dell'organizzazione mondiale del turismo (unwto).
# Devo tenermi soltanto gli Stati che ho nel dataset delle conferenze

df_countries_unwto = pd.read_csv('tourist_arrivals_by_country_and_year.csv')

def edit(x):
    return x.lower().title() # UNITED KINGDOM ---> United Kingdom

df_countries_unwto['Paesi'] = df_countries_unwto['Paesi'].map(edit)

print(list(df_countries_unwto['Paesi']))
```

```python
"""
Ci sono alcuni stati presenti in tutte e tre le liste, ma sotto diverso nome.
Esempio: Russia / Russian Federation. Controllo quindi quali e quanti elementi ci sono in 
comune (e non) tra la prima lista(cioè quella con i luoghi delle conferenze)
e le altre due, in modo da fare poi le correzioni opportune. """

elements_in_common_countries_wef = [x for x in countries if x in list(df_countries_WEF['Stato']) ]
elements_in_common_countries_unwto = [x for x in countries if x in list(df_countries_unwto['Paesi'])]

print(elements_in_common_countries_wef)
print(len(elements_in_common_countries_wef))
print(elements_in_common_countries_unwto)
print(len(elements_in_common_countries_unwto))
```

```python
elements_not_in_common_countries_wef = [x for x in countries if x not in list(df_countries_WEF['Stato']) ]
print('Paesi delle conferenze che non sono nel dataset WEF: \n' + str(elements_not_in_common_countries_wef))

elements_not_in_common_wef_countries = [x for x in list(df_countries_WEF['Stato']) if x not in countries ]
print('\nPaesi del dataset WEF che non sono nel dataset delle conferenze: \n' + str(elements_not_in_common_wef_countries))

"""
Czechia = Czech Republic
Iran = Iran, Islamic Rep.
Russia = Russian Federation
Slovakia = Slovak Republic
South Korea = Korea, Rep.
Taiwan = Taiwan, China
Vietnam = Viet Nam

Correggo (a mano) i valori della seconda lista, in modo che il join riesca a matchare bene
"""

df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Czech Republic'] = 'Czechia'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Iran, Islamic Rep.'] = 'Iran'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Russian Federation'] = 'Russia'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Slovak Republic'] = 'Slovakia'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Korea, Rep.'] = 'South Korea'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Taiwan, China '] = 'Taiwan'
df_countries_WEF['Stato'][df_countries_WEF['Stato']== 'Viet Nam'] = 'Vietnam'
```

```python
# faccio la stessa cosa, ma con il dataset dell'unwto
elements_not_in_common_countries_unwto = [x for x in countries if x not in list(df_countries_unwto['Paesi']) ]
print('Paesi delle conferenze che non sono nel dataset UNWTO: \n' + str(elements_not_in_common_countries_unwto))

elements_not_in_common_unwto_countries = [x for x in list(df_countries_unwto['Paesi']) if x not in countries ]
print('\nPaesi del dataset UNWTO che non sono nel dataset delle conferenze: \n' + str(elements_not_in_common_unwto_countries))

"""
Bolivia = Bolivia, Plurinational State Of
Bosnia and Herzegovina = Bosnia And Herzegovina
Czechia = Czech Republic
Iran = Iran, Islamic Republic Of
Moldova = Moldova, Republic Of
Russia = Russian Federation
Saint Kitts and Nevis = Saint Kitts And Nevis
South Korea = Korea, Republic Of
Taiwan = Taiwan Province Of China
Tanzania = Tanzania, United Republic Of
The Bahamas = Bahamas
United States = United States Of America
Venezuela = Venezuela, Bolivarian Republic Of
Vietnam = Viet Nam

Anche qui correggo a mano
"""
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Bolivia, Plurinational State Of'] = 'Bolivia'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Bosnia And Herzegovina'] = 'Bosnia and Herzegovina'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Czech Republic'] = 'Czechia'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Iran, Islamic Republic Of'] = 'Iran'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Moldova, Republic Of'] = 'Moldova'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Russian Federation'] = 'Russia'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Saint Kitts And Nevis'] = 'Saint Kitts and Nevis'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Korea, Republic Of'] = 'South Korea'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Taiwan Province Of China'] = 'Taiwan'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Tanzania, United Republic Of'] = 'Tanzania'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Bahamas'] = 'The Bahamas'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'United States Of America'] = 'United States'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Venezuela, Bolivarian Republic Of'] = 'Venezuela'
df_countries_unwto['Paesi'][df_countries_unwto['Paesi'] == 'Viet Nam'] = 'Vietnam'

list(df_countries_unwto['Paesi'])
```

```python
# unisco tutti gli indici relativi agli Stati in un unico dataset
df['Country'] = df['ConferenceLocation'].str.split(',').str[-1].str.lstrip()

df_tourist_arrivals = df.merge(df_countries_unwto, left_on='Country', right_on='Paesi', how='left')
df_indexes_by_country = df_tourist_arrivals.merge(df_countries_WEF, left_on='Country', right_on='Stato', how='left')
del df, df_countries_unwto, df_countries_WEF
df_indexes_by_country.drop(columns=['Paesi', 'Stato'], inplace=True)
df_indexes_by_country

"""
NB. gli arrivi di turisti sono espressi in migliaia
"""
```

```python
df_indexes_by_country.drop_duplicates(inplace=True)
df_indexes_by_country.to_csv('all_indexes_by_country.csv', index=False)
```
