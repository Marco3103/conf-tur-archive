```python
import PyPDF2
from PyPDF2 import PdfReader
import pandas as pd
import os
```

```python
reader = PdfReader('WEF_TTCR_2019.pdf')
```

```python
indexes = ['TTCI', 'EE', 'TTPEC','INF','NCR', 'BE', 'SS', 'HH', 'HRLM', 'ICTR','PTT', 'IO', 'PC', 'ES', 'ATI', 'GPI', 'TSI', 'NR', 'CRBT']
df_final = pd.DataFrame(columns= ['Stato', 'TTCI', 'EE', 'TTPEC','INF','NCR', 'BE', 'SS', 'HH', 'HRLM', 'ICTR','PTT', 'IO', 'PC', 'ES', 'ATI', 'GPI', 'TSI', 'NR', 'CRBT'])
df_final 
```

TTCI : The Travel & Tourism Competitiveness Index 2019 (la media aritmetica di tutti gli altri)  
EE : Enabling Environment  
TTPEC : Travel & Tourism Policy and Enabling Conditions  
INF : Infrastructure  
NCR : Natural and Cultural Resources  
BE : Business Environment  
SS : Safety and Security  
HH : Health and Hygiene  
HRLM : Human Resources and Labour Market  
ICTR : ICT Readiness  
PTT : Prioritization of Travel & Tourism  
IO : International Openness  
PC : Price Competitiveness  
ES : Environmental Sustainability  
ATI : Air Transport Infrastructure  
GPI : Ground and Port Infrastructure  
TSI : Tourist Service Infrastructure  
NR : Natural Resources  
CRBT : Cultural Resources and Business Travel  

Fonte : https://reports.weforum.org/travel-and-tourism-competitiveness-report-2019/downloads/
alla voce "full report PDF"  
Tutti i valori sono compresi tra 1 e 7

```python
for page in list(range(80,99)): # sono le pagine del pdf da cui estrarre il testo
    page_content = reader.pages[page]
    print(page_content.extract_text())
    """ Il contenuto è troppo sporco per pulirlo con del codice. Salvo in un file, tolgo a mano 
        ciò che non serve e lo riapro, per poi automatizzare l'estrazione dei dati necessari"""
    f = open(f'text_to_clean_{page}.txt', 'w')
    f.write(page_content.extract_text())
    f.close()
```

```python
for index,page in zip(indexes, list(range(80,99))):
    df = pd.DataFrame(columns=['Stato', f'{index}'])
    with open(f'text_to_clean_{page}.txt', 'r') as f:
        while True:
            df = df.append({'Stato': f.readline().replace('\n', ''), f'{index}': f.readline().replace('\n', '')}, ignore_index=True)
            f.readline()
            if f.readline()=='': # controllo se sono alla fine del file
                break
                
    os.remove(f'text_to_clean_{page}.txt')
    df = df.sort_values(by=['Stato']).reset_index(drop=True)

    if page==80: # alla prima iterazione inserisco gli Stati nel df finale
        df_final['Stato'] = df['Stato']
        df_final= df_final.reset_index(drop=True)

    df_final[f'{index}'] = df[f'{index}']
    
```
