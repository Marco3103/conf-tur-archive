```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import requests
import re
```

```python
df = pd.read_csv('out_citations_and_conferences_location_ready.csv')

citta = df['ConferenceLocation'].str.split(',').str[0].unique()
```

```python
# creo una Series vuota, dove andrà a finire il numero di risultati per ogni città
values = pd.Series(np.nan, index= [x for x in range(len(citta))])
d = {'citta': citta, 'risultati': values}
google_results = pd.DataFrame(d)
google_results
```

```python
#scraping dei risultati della ricerca su Google, per ogni città

def start_soup(link):
    response = requests.get(link, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


# senza l'header mi restituisce la pagina di accettazione cookie, quindi non riesco a estrarre il numero di risultati
headers = {
    'User-agent':
    'Mozilla/5.0 (X11; Linux x86_64; rv:99.0) Gecko/20100101 Firefox/99.0' }

for x in citta:
    print(x)
    base_link = f"https://www.google.com/search?q={x}"

    try:
        results = start_soup(base_link).find("div", {"id": "result-stats"}).text
        # estrazione del numero dal tag HTML
        reg = re.search('Circa (.*) risultati', results).group(1)
        print(reg)
        # inserisco il numero nel dataframe, in corrispondenza della città
        google_results['risultati'] = np.where(google_results['citta']!=x, google_results['risultati'], reg)
    except:
        google_results['risultati'] = np.where(google_results['citta']!=x, google_results['risultati'], np.nan)
google_results
```

```python
"""osservazioni:
    1.Non tutti i luoghi presi dal file con le citazioni sono città. Esempio : Veneto, Italy
    2. alcuni valori sono NaN, perchè quando si effettua la ricerca di quella città, Google
     non restituisce il numero di risultati, bensì altro
   """
print(google_results[google_results['risultati'].isna()])
```
