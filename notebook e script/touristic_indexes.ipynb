```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('out_citations_and_conferences_location_ready.csv')
# estraggo lo Stato dalla colonna delle location, per vedere quanti e quali sono

stati = df['ConferenceLocation'].str.split(',').str[-1]
print(stati.unique())
print('\n Ho ' + str(len(stati.unique())) + ' Stati diversi tra loro')
```

```python
"""Avendo 126 Stati, i dataset dell'OECD non sono abbastanza utili, in quanto contengono dati di circa 60 Paesi
Procedo quindi con i dataset dell'Organizzazione mondiale del turismo.
Estraggo i dati dai file excel. Tra i dati più coerenti e rilevanti ci sono gli arrivi di turisti
""" 

paesi = pd.read_excel('unwto-inbound-arrivals-data.xlsx', usecols=[3])
paesi.drop([0,1], inplace=True)
# mantengo solo le righe con i nomi degli Stati
paesi = paesi.iloc[::6]
paesi.drop([1340,1346], inplace=True)

"""
Creazione del dataframe con i paesi e i rispettivi arrivi di turisti negli anni.
Importo solo le colonne utili, poi taglio le righe non necessarie.
NB. i numeri di turisti sono espressi in migliaia
"""

arrivi = pd.read_excel('unwto-inbound-arrivals-data.xlsx', header=2, usecols=range(11,37))
arrivi = arrivi.iloc[2::6]
arrivi.drop([1340], inplace=True)
arrivi.replace('..',np.NaN, inplace=True)
arrivi
```

```python
# aggiungo gli Stati al dataframe dei valori
arrivi['Paesi'] = paesi

# La colonna dei Paesi è in fondo al dataframe, la imposto come prima colonna per una miglior leggibilità
cols = arrivi.columns.tolist()
cols = cols[-1:] + cols[:-1]
arrivi = arrivi[cols]
arrivi
```

```python
arrivi.to_csv('tourist_arrivals_by_country_and_year.csv', index=False)
```
